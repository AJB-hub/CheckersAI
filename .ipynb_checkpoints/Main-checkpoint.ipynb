```python
import pandas as pd
import numpy as np
import random as rand
```

```python
state_zero = [['w','','w','','w','','w',''],
              ['','w','','w','','w','','w'],
              ['w','','w','','w','','w',''],
              ['','','','','','','',''],
              ['','','','','','','',''],
              ['','b','','b','','b','','b'],
              ['b','','b','','b','','b',''],
              ['','b','','b','','b','','b']
             ]
```

```python
def resetBoard():
    global board_state
    board_state = pd.DataFrame(state_zero,index=range(0,8),columns=range(0,8))
    return board_state
resetBoard()
```

```python
# Finds the location of pieces on a board

def pieceFinder(color):
    pieceLoc = []
    for column in np.arange(0,8,1):
        for row in np.arange(0,8,1):
            if board_state.iloc[row,column] == color:
                pieceLoc += [[row,column]]
    return pieceLoc
```

```python
def blackMove():
    pieceCount = len(pieceFinder('b'))
    isTurn = True

    while isTurn == True:
        for piece in np.arange(0,pieceCount,1):
            row = pieceFinder('b')[piece][0]
            column = pieceFinder('b')[piece][1]
            try: 
                # Right diagonal movement
                if board_state.iloc[row-1,column+1] == '':
                    board_state.iloc[row-1,column+1] = 'b'
                    board_state.iloc[row,column] = ''
                    display(board_state)
                    isTurn = False
                    break
                # Left diagonal movement
                elif board_state.iloc[row-1,column-1] == '':
                    board_state.iloc[row-1,column-1] = 'b'
                    board_state.iloc[row,column] = ''
                    display(board_state)
                    isTurn = False
                    break
                # if the left diagonal square has a white piece and the diagonal after is empty, jump and capture the piece
                elif board_state.iloc[row-1,column+1] == 'w' & board_state.iloc[row-2,column+2]=='':
                    board_state.iloc[row-2,column+2] = 'b'
                    board_state.iloc[row,column] = ''
                    board_state.iloc[row-1,column+1] = ''
                    display(board_state)
                    isTurn = False
                    break
                #right diagonal
                elif board_state.iloc[row-1,column-1] == 'w' & board_state.iloc[row-2,column-2]=='':
                    board_state.iloc[row-2,column-2] = 'b'
                    board_state.iloc[row,column] = ''
                    board_state.iloc[row-1,column+1] = ''
                    display(board_state)
                    isTurn = False
                    break
                      
                else:
                    print("Piece unable to move\n")


            except: 
                print("Cannot Find a Legal Move.")
```

```python
def whiteMove():
    pieceCount = len(pieceFinder('w'))
    isTurn = True
    while isTurn == True:
#         Find a Piece on the board to move
        for piece in np.arange(0,pieceCount,1):
            row = pieceFinder('w')[piece][0]
            column = pieceFinder('w')[piece][1]
#             Implement a method of selecting a strategy after a piece has been selected
            try: 
                elif board_state.iloc[row+1,column+1] == '':
                    board_state.iloc[row+1,column+1] = 'w'
                    board_state.iloc[row,column] = ''
                    display(board_state)
                    isTurn = False
                    break

                elif board_state.iloc[row+1,column-1] == '':
                    board_state.iloc[row+1,column-1] = 'w'
                    board_state.iloc[row,column] = ''
                    display(board_state)
                    isTurn = False
                    break
                # if the left diagonal square has a white piece and the diagonal after is empty, jump and capture the piece
                elif board_state.iloc[row+1,column+1] == 'b' & board_state.iloc[row+2,column+2]=='':
                    board_state.iloc[row+2,column+2] = 'w'
                    board_state.iloc[row,column] = ''
                    board_state.iloc[row+1,column+1] = ''
                    display(board_state)
                    isTurn = False
                    break
                #right diagonal
                elif board_state.iloc[row+1,column-1] == 'b' & board_state.iloc[row+2,column-2]=='':
                    board_state.iloc[row+2,column-2] = 'w'
                    board_state.iloc[row,column] = ''
                    board_state.iloc[row+1,column-1] = ''
                    display(board_state)
                    isTurn = False
                    break

                else:
                    print("Piece unable to move\n")


            except: 
                print("Cannot Find a Legal Move.")
```

```python

resetBoard()
for i in np.arange(0,3,1):
    if i % 2 == 0:
        blackMove()
    else:
        whiteMove()
        
```
